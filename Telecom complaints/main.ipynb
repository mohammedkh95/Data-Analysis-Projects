```python
# ROW PRE-PROCESSING

# Full Name
df['Full Name'] = df['Full Name'].str.strip().str.title()

# Signup date
df['Signup Date'] = pd.to_datetime(df['Signup Date'], errors='coerce')
df['Signup Date'] = df['Signup Date'].dt.strftime('%B %d, %Y')

# Last Transaction
df['Last Transaction'] = df['Last Transaction'].str.strip()
df['Last Transaction'] = pd.to_datetime(df['Last Transaction'], format='mixed')
df['Last Transaction'] =df['Last Transaction'].astype(str)
df[['Transaction Date','Transaction Time']] = df['Last Transaction'].str.split(' ', n=1, expand=True)
# ----
df['Transaction Date'] = pd.to_datetime(df['Transaction Date']).dt.strftime('%B %d, %Y')
df['Transaction Time'] = pd.to_datetime(df['Transaction Time'], format='%H:%M:%S').dt.strftime('%I:%M %p')

# Plan Type
df['Plan Type'] = df['Plan Type'].str.strip().str.capitalize()
plan_type = {'Bronze':'bronze', 'Bronze plan':'bronze', 'Gold plan':'gold', 'GOLD':'gold', 'Silver':'silver', 'SILVER':'silver','Gold':'gold'}
df['Plan Type'] = df['Plan Type'].map(plan_type).str.capitalize()

# Monthly fee
df['Monthly Fee'] = df['Monthly Fee'].str.replace(r'[@#$%&]','', regex=True)
df['Monthly Fee'] = df['Monthly Fee'].str.replace(r'USD','', regex=True).str.strip()
df['Monthly Fee'] = pd.to_numeric(df['Monthly Fee'], errors='coerce')

# Complaint category
df['Complaint Category'] = df['Complaint Category'].str.strip()
complaint_cat={'Technical':'technical support', 'Billing':'billing', 'billing issue':'billing', 'customer service':'customer service', 'Tech Support':'technical support', 'Service':'other service'}

df['Complaint Category'] = df['Complaint Category'].map(complaint_cat).str.capitalize()

# Satisfaction rating
df['Satisfaction Rating'] = df['Satisfaction Rating'].str.strip()
sat_rating = {'‚≠êÔ∏è‚≠êÔ∏è‚≠êÔ∏è‚≠êÔ∏è‚≠êÔ∏è':'5', '‚≠êÔ∏è‚≠êÔ∏è':'2','‚≠êÔ∏è':'1', '‚≠êÔ∏è‚≠êÔ∏è‚≠êÔ∏è‚≠êÔ∏è':'4','‚≠êÔ∏è‚≠êÔ∏è‚≠êÔ∏è':'3', 'four':'4', 'five':'5', 'three':'3', 'two':'2', 'one':'1'}
df['Satisfaction Rating'] = df['Satisfaction Rating'].map(sat_rating)

# Is churned
df['Is Churned'] = df['Is Churned'].str.strip()
churn_rating = {'1':'yes','True':'yes','false':'no','YES':'yes','0':'no','No':'no'}
df['Is Churned'] =df['Is Churned'].map(churn_rating).str.capitalize()

# Address
df['Address'] = df['Address'].str.strip()
df[['Street Address','City','State']] = df['Address'].str.split(',',n=2,expand=True)
df[['State', 'ZIP Code']] = df['State'].str.strip().str.split(' ', n=1, expand=True)

# Fill missing
df = df.fillna('Unavailable')
```

```python
# COLUMN PROCESSING

# Strip spaces
df.columns = df.columns.str.strip()

# Drop column
df = df.drop(columns=['Last Transaction','Customer ID','Address'])

# Rename columns
df = df.rename(columns={'Monthly Fee': 'Monthly Fee ($USD)','Complaint Category':'Complaint Type','Satisfaction Rating':'Rating', 'Is Churned':'Churned'})

# Reorder column
column_order = [
    'Full Name', 'Street Address', 'City', 'State', 'ZIP Code',
    'Signup Date', 'Plan Type', 'Monthly Fee ($USD)',
    'Transaction Date', 'Transaction Time',
    'Complaint Description', 'Complaint Type', 'Rating', 'Churned',
    'Notes'
]

# Reindex
df = df.reset_index(drop = True)
df = df[column_order]
```

```python
# Convert to excel
df.to_excel('test.xlsx', index=False)
```
